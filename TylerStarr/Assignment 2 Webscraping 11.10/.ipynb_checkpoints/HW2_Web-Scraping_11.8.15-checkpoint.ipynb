***make_soup ***
Takes a URL, makes it into Beautiful Soup, and returns the URL



```python
def make_soup(url):
    page = urllib2.urlopen(str(url)).read()
    soup = BeautifulSoup(page)
    soup.prettify()
    return soup
```

***get_cols(soup)*** 
Retrieves the names of the header columns given a table element
Takes a Table Element
Returns Names of the header columns

```python
test_soup = make_soup(str('http://espn.go.com/nba/statistics/player/_/stat/scoring-per-game/sort/avgPoints/qualified/false'))

def get_cols(soup):
    #print "This is get_cols"
    table_div = soup.find(id='my-players-table')
    #print table_div.prettify()[0:4000]
    table = table_div.find("table")
    #print table
    table_head = table.find(attrs={"class":'colhead'})
    #print table_head.prettify()
    header_cols = table_head.findAll('td')
    #print header_cols
    
    #NOW GOING INTO LOOP TO GET ALL HEADER COLS
    cols = []
    for header_col in header_cols:
        val = header_col.string
        if val != 'RK':
            cols.append(val)
        if val == 'PLAYER':
            cols.append('POSITION')
    #print cols
    return cols


```

```python
cols = get_cols(test_soup)
print "What!", cols
```

***Get_data***
Retrieves the actual table datagiven a table
Takes the made soup URL and the columns
Returns a dictionary of the data

```python


def get_data(soup, cols):
    table_div = soup.find(id='my-players-table')
    #print table_div.prettify()[0:4000]
    table = table_div.find("table")
    table_rows = table.findAll('tr')
    #print table_rows
    players_stats_dicts = []
    for row in table_rows:
        if row.attrs['class'][0]== 'colhead':
            continue
        player_stats = []
        row_cols = row.find_all('td')
        col_vals = []
        player_col = row_cols[1]
        player_name = player_col.find('a').string
        player_position = player_col.contents[1]
        player_position = player_position.split(' ')[1]
        player_stats.append(player_name)
        player_stats.append(player_position)
        for i in range(2, len(row_cols)):
            stat = row_cols[i].string
            player_stats.append(stat)
        player_stats = dict(zip(cols, player_stats))
        players_stats_dicts.append(player_stats)
    #print players_stats_dicts[0:5]
    return players_stats_dicts    
        
one_dict = get_data(test_soup, cols)
```

***url_changer(url)***
Takes a URL and makes an array of all the urls that exist for the Score and Assist pages
Takes a single URL
Returns an array of all the URLs

```python
def url_changer(url):
    new_url = str(url) + "/count/"
    all_urls = []
    counter = 0
    for i in range(11):
        counter = (i*40) + 1
        #print i, counter
        i += 1
        temp_url = new_url + str(counter)
        all_urls.append(temp_url)
        #print all_urls
    #print all_urls
    return all_urls

all_scoring_urls = url_changer("http://espn.go.com/nba/statistics/player/_/stat/scoring-per-game/sort/avgPoints/qualified/false")
#print all_scoring_urls
```

```python
def run_through(all_urls):
    #print all_urls
    all_data = []
    for url in all_urls:
        souped_url = make_soup(str(url))
        cols = get_cols(souped_url)
        #print cols
        data_1 = get_data(souped_url, cols)
        #print data_1
        for player in data_1:
            all_data.append(player)
            #print all_data
    
    #print all_data
    return all_data

ex_dict = run_through(all_scoring_urls)
```

```python
def real_thing():
    print "Let's do it!"
    #SCORING
    score_url = 'http://espn.go.com/nba/statistics/player/_/stat/scoring-per-game/sort/avgPoints/qualified/false'
    
    all_urls = url_changer(score_url)
    #print all_urls
    dict_data = run_through(all_urls)
    #print dict_data
    score_df = turn_into_df(dict_data)
    #print score_df
    
    
    #ASSISTS
    assist_url = 'http://espn.go.com/nba/statistics/player/_/stat/assists/qualified/false'
    all_urls = url_changer(assist_url)
    #print all_urls
    dict_data = run_through(all_urls)
    #print dict_data
    assist_df = turn_into_df(dict_data)
    #print assist_df
    
    #JOINING
    #joined_df = pd.DataFrame.join(assist_df, score_df, on = "PLAYER", lsuffix = " ")
    #joined_df = pd.DataFrame.join(assist_df, score_df, lsuffix = " ")
    joined_df = pd.merge(assist_df, score_df, on='PLAYER', how='right')
    
    joined_df.head()
    #print joined_df
    return joined_df
    
    
joined_df = real_thing()
joined_df.head()
```
